```python
#This works for rivals. I'll copy it for a nemesis just to make sure I don't break it

# I think the dudes with spells, and the complicated ones like the Dreadgazer, need to be done manually
folder = "C:\Users\dipic\Google Drive\Gaming\Pen and Paper RPGs\Genesys\Adventure Design\\"
#I will use this to keep track of all of the abilities with full text to see what I can replace
All_Abilities_List = []
All_Abilities_List_name_only = []
JSONtext = ''

char_list = ["Brawn","Agility","Intellect","Cunning","Willpower","Presence"]
    
AdversaryText = open(folder + "Official Adversary Text.txt","r")

#Start at 999999 so when I use the initial between it's always false, but then after we set the ability
#line it works
ability_line_num = 999999
skill_line_num = 999999
talent_line_num = 999999
spell_line_num = 999999
equipment_line_num = 999999
gear_line_num = 999999

for line_num,line_full in enumerate(AdversaryText):
    line = line_full
    #Firstly, I want to replace all tabs and double-spaces with single spaces to make parsing easier
    line = line_full.replace('\t', ' ')
    while line.find("  ") > -1:
        line = line.replace('  ', ' ')
    #I also want to remove trailing spaces from lines
    line = line.replace(' \n','\n')
    
    #Grab the setting
    if line.replace('\n','') in ['Fantasy', 'Steampunk', 'Science Fiction', 'Space Opera', 'Age of Myth', 'Modern', 'Monster World', 'Post Apolocalypse', 'Weird War', 'General']:
        setting = line.replace('\n','')
    
    #First line is the Power Level
    if line[:line.find(":")] == "Power Level":
        power_levels = line[line.find(":")+2:].split(",")
        CPL = power_levels[0]
        SPL = power_levels[1]
        GPL = power_levels[2].replace('\n','')
    
    #For the second line, we know it starts with the name and type
    #print('Line Number: ' + str(line_num))
    #Print where I expect the characteristic text to be
    #print(line[:line.find(":")] in char_list)
    if line.find("(Minion)") > -1 or line.find("(Rival)") > -1 or line.find("(Nemesis)") > -1:
        #Grab the character name
        JSONtext += '\t{\n\t\t"name": "'
        p1 = line.find(" (")
        JSONtext += line[:p1] + '",\n\t\t"type": "'

        #Grab the type
        p2 = line.find(")")
        adversary_type = line[p1+2:p2]
        #This file doesn't have the descriptions
        #JSONtext += adversary_type + '",\n\t\t"description": "'
        JSONtext += adversary_type + '",\n\t\t"characteristics": {'
        
        #Loop through characteristics
        for c in char_list:
            #print(line.find(c))
                if c == "Presence":
                    JSONtext += '\n\t\t\t"' + c + '": ' + '' + '\n\t\t},\n\t\t"derived": {'
                else:
                    JSONtext += '\n\t\t\t"' + c + '": ' + '' + ','
            
        #Loop through derived characteristics (for rivals at least)
        for d in ["Soak","Wounds", "Strain", "M/R Defense"]:

                if d == "M/R Defense":
                    #I'm assuming the defence will always be single digits
                    JSONtext += '\n\t\t\t"defence": [' + '' + ',' + '' + ']\n\t\t},\n\t\t"skills": '
                elif not(d == "Strain" and adversary_type == "Minion"):
                    JSONtext += '\n\t\t\t"' + d.lower() + '": ' + '' + ','
                    
    #This is to make sure I start with the correct bracket
    if line[:line.find(":")] == "Skills":
        if adversary_type == "Minion":
            JSONtext += '['
        else:
            JSONtext += '{'
    #I need to handle Melee (X) skills and knowledge skills
    if line[:line.find(":")] == "Skills":
        skill_line_num = line_num
        
    #This needs to handle the next few lines of skills
    if line_num >= skill_line_num and line[:line.find(":")] != "Talents":
        #Skills: Athletics 1, Brawl 2, Resilience 2, Melee (Light) 3
        SkillSplit = line.replace('\n','').replace('Skills: ', '').split(", ")
        for s in SkillSplit:
            #Minion skills have no number
            #print s
            #I need the replace so that Melee (Light) 2 becomes "Melee: Light": 2
            if adversary_type == "Minion":
                JSONtext += '"' + s.replace(" (", ": ").replace(")","") + '",'
            else:
                #I'm assuming the skills will always be single digits
                JSONtext += '\n\t\t\t"' + s.replace(" (", ": ").replace(")","").replace(',','')[:-2] + '": ' + s.replace(",", "")[-1:] + ","
        #Now take off that final comma
        #I should check here for talents, probably, but let's go one example at a time
        JSONtext = JSONtext[:len(JSONtext)-1]
        
    #This is to make sure I end with the correct bracket
    if line[:line.find(":")] == "Talents":
        skill_line_num = 999999
        if adversary_type == "Minion":
            JSONtext += '],\n\t\t"talents": ['
        else:
            JSONtext += '\n\t\t},\n\t\t"talents": ['
                  
    if line[:line.find(":")] == "Talents":
        talent_line_num = line_num
        
    #This needs to handle the next few lines of talents. Always one per line
    if line_num > talent_line_num and line[:line.find(":")] != "Abilities" and line != "\n":
        #print line
        #JSONtext += '\n\t\t\t{"name": "' + line[:line.find(":")-1] + '", "description": "' + \
        #                  line[line.find(":")+1:] + '"}, '
        JSONtext += '"' + line[:line.find(":")] + '",'
                  
    if line[:line.find(":")] == "Abilities":
        talent_line_num = 999999
        #Now take off that final comma from the talents
        JSONtext = JSONtext[:len(JSONtext)-1] + '],\n\t\t"abilities": ['
        ability_line_num = line_num
        
    if line[:line.find(":")] == "Spells":
        spell_line_num = line_num
        
    #This needs to handle the next few lines of abilities
    if line_num > ability_line_num and line[:line.find(":")] != "Equipment" and line != "\n":
        #print line
        #I think this split is pointless as they are always on separate lines
        #AbilitySplit = line.replace('\n','').replace('Abilities: ', '').split(", ")
        Ability = line.replace('\n','')
        if line_num > spell_line_num:
            Ability = 'Spells| ' + Ability
        #print Ability
        #If the ability has no description in brackets, it must be a generic one
        if Ability.find(":") == -1:
            JSONtext += '"' + Ability + '", '
        else:
            #Keep track of all of the abilities to see what I can make generic
            All_Abilities_List.append(Ability)
            All_Abilities_List_name_only.append(Ability[:Ability.find(":")])
            JSONtext += '\n\t\t\t{"name": "' + Ability[:Ability.find(":")] + '", "description": "' + \
                                Ability[Ability.find(":")+2:] + '"}, '
        
    if line[:line.find(":")] == "Equipment" and line[:line.find(":")] != "Source":
        ability_line_num = 999999
        spell_line_num = 999999
        equipment_line_num = line_num
        #Now take off that final comma and space from the abilities
        JSONtext = JSONtext[:len(JSONtext)-2] + '],\n\t\t"weapons": ['
        
    if line_num > equipment_line_num:
        #print line
        #WeaponSplit = line.replace('\n','')[line.find(":")+2:].split("), ")
        ##Trying to keep the delimiters
        #d = "), "
        #To make this delimiting easier
        #newlinetext = line.replace('\n','')[line.find(":")+2:] + ", "
        ##print newlinetext
        #WeaponSplit =  [e+")" for e in newlinetext.split("), ")]
        #for w in WeaponSplit:
        #One line per weapon
            #First, make sure it's a weapon
            #print w
        if line.find("Damage") > -1:
            #print line
            JSONtext += '\n\t\t\t{\n\t\t\t\t"name": "' + line[:line.find(":")] + '",\n\t\t\t\t'
            WeaponStatSplit = line[line.find(":")+1:].split("; ")
            #First is skill
            JSONtext += '"skill": "' + WeaponStatSplit[0][1:] + '",\n\t\t\t\t'
            #Then the damage
            JSONtext += '"damage": ' + WeaponStatSplit[1][WeaponStatSplit[1].find(" ")+1:] + ',\n\t\t\t\t'
            #Then the critical rating
            JSONtext += '"critical": ' + WeaponStatSplit[2][WeaponStatSplit[2].find(" ")+1:] + ',\n\t\t\t\t'
            #Then the range
            JSONtext += '"range": "' + WeaponStatSplit[3].replace('Range ', '').replace('(','').replace(')','') + '",\n\t\t\t\t'
            #Now get the special qualities. Get rid of that trailing bracket
            #I need to handle weapons without qualities, here
            if len(WeaponStatSplit) >= 5:
                Qualities = WeaponStatSplit[4].split(", ")
                JSONtext += '"qualities": ['
                for q in Qualities:
                    #print q.replace("\n",'')
                    JSONtext += '"' + q.replace("\n",'') + '",'
                #Now take off that final comma after the last quality
                JSONtext = JSONtext[:len(JSONtext)-1] + ']\n\t\t\t},'
            else:
                JSONtext += '},'
            
        #This is for the generic gear
        elif line.find("Damage") == -1 and gear_line_num == 999999:
            gear_line_num = line_num
            JSONtext = JSONtext + '\n\t\t],\n\t\t"gear": ['
            
        if line_num >= gear_line_num and line.find("Source") == -1:
            JSONtext = JSONtext + '"' + line.replace('\n','') + '",'
            
        #Rid of the final comma from the gear, too
        if line_num >= gear_line_num and line.find("Source") > -1:
            JSONtext = JSONtext[:-1]

        #Finish off the gear and add the tags
        if line[:line.find(":")] == "Source":
            equipment_line_num = 999999
            gear_line_num = 999999
            source = line[line.find(":")+2:].replace('\n','')
            JSONtext = JSONtext + '],\n\t\t"tags": ["source:' + source[:source.find(' p.')] + '", "Official", ' + \
                            '"setting:' + setting + '", "Combat CR:' + CPL + '", "Social CR:' + SPL + \
                            '", General CR:' + GPL + '"]\n\t},\n'
                
        #JSONtext = JSONtext[:len(JSONtext)-1] + '\n\t\t],\n\t\t"tags": [' + \
        #                '"source:Creature Catalogue", "Official", "Fantasy", ' + \
        #                '"Combat CR:TBD", "Social CR:TBD", "General CR:TBD"]\n\t},\n'
    
    #if line_num >= 18:
    #    break

from collections import Counter

print 'Duplicate abilities with exact text match:'
for a in Counter(All_Abilities_List):
    if Counter(All_Abilities_List)[a] > 1:
        print a
        
print 'Duplicate abilities with name match:'
for a in Counter(All_Abilities_List_name_only):
    if Counter(All_Abilities_List_name_only)[a] > 1:
        print a
        
#After going through all of that, remove the final ,
JSONtext = JSONtext[:-2]
#print JSONtext
AdversaryText.close()

JSONformatFile = open(r"AdversaryText JSON.json","w+")
JSONformatFile.write(JSONtext)
JSONformatFile.close()
```
